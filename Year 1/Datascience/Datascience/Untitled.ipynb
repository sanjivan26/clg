```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
plt.title("rand")
plt.scatter(np.random.rand(20),np.random.rand(20),c=np.random.rand(20),cmap='rainbow')
plt.colorbar()
plt.show()
```

```python
x=np.linspace(0,10,10)
plt.title("lol")
plt.plot(x,np.random.rand(10),label='stonks')
plt.plot(x,np.random.rand(10),label='stonk\'nt')
plt.legend()
plt.show()
```

```python
x=['a','b','c','d']
ex=[.1,.1,.1,.2]
plt.pie(np.random.rand(4),labels=x,explode=ex,shadow=True)
plt.legend(title="")
plt.show()
```
