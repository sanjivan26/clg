## Read Total profit of all months and show it using a line plot

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
dataset=pd.read_csv('company_sales_data.csv')
df=pd.DataFrame(dataset)
```

```python
df
```

```python
df.total_profit
```

```python
plt.plot(df.index,df.total_profit)
```

## 2. Get total profit of all months and show line plot with the following Style properties
Generated line plot must include following Style properties: –
•	Line Style dotted and Line-color should be red
•	Show legend at the lower right location.
•	X label name = Month Number
•	Y label name = Sold units number
•	Add a circle marker.
•	Line marker color as read
•	Line width should be 3


```python
plt.plot(df.month_number,df.total_profit,linestyle='dotted',linewidth='3',color='r',marker='o')
plt.xlabel('Month Number')
plt.ylabel('Sold units number')
plt.legend(['profit over months'],loc='lower right')
```

## 3: Read all product sales data and show it  using a multiline plot

```python
for i in range(1,7):
    plt.plot(df.month_number,df.iloc[:,i],label=df.columns[i])
    plt.xlabel('Month')
    plt.ylabel('sales')
    plt.legend()
```

##  4: Read toothpaste sales data of each month and show it using a scatter plot

```python
plt.scatter(df.month_number,df.toothpaste)
```

## Read face cream and facewash product sales data and show it using the bar chart

```python
plt.bar(df.month_number-0.13,df.facecream,width=0.26,label='facecream')
plt.bar(df.month_number+0.13,df.facewash,width=0.26,label='facewash')
plt.legend()
```

## Calculate total sale data for last year for each product and show it using a Pie chart

```python
values=[]
for i in range(1,7):
    values.append(sum(df.iloc[:,i]))
plt.pie(values,labels=df.columns[1:7])
plt.show()
```
